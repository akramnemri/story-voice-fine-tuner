```python
from transformers import pipeline, GPT2Tokenizer
import torch
import gradio as gr

# Load directly from Google Drive (permanent location)
model_path = "/content/drive/MyDrive/story_voice_finetuned"

tokenizer = GPT2Tokenizer.from_pretrained(model_path)
tokenizer.pad_token = tokenizer.eos_token

generator = pipeline(
    "text-generation",
    model=model_path,
    tokenizer=tokenizer,
    device=0 if torch.cuda.is_available() else -1
)

print("✅ Model loaded successfully from Google Drive!")
```

```python
def generate_text(prompt, max_new_tokens=200, temperature=0.85, top_k=50, top_p=0.95):
    sequences = generator(
        prompt,
        max_new_tokens=max_new_tokens,
        num_return_sequences=1,
        do_sample=True,
        temperature=temperature,
        top_k=top_k,
        top_p=top_p,
        pad_token_id=tokenizer.pad_token_id
    )
    return sequences[0]['generated_text']
```

```python
with gr.Blocks(title="Story-Voice Fine-Tuner", theme=gr.themes.Soft()) as demo:
    gr.Markdown("# Story-Voice Fine-Tuner\nYour personal AI coauthor")

    with gr.Row():
        with gr.Column(scale=2):
            prompt_input = gr.Textbox(
                label="Start your scene",
                placeholder="He reached the entrance and blasted the door open in a surge of rage...",
                lines=3
            )
            max_new = gr.Slider(50, 400, value=200, step=10, label="Max new tokens")
            temp = gr.Slider(0.1, 1.3, value=0.85, step=0.05, label="Temperature (creativity)")
            topk = gr.Slider(10, 100, value=50, step=5, label="Top-K")
            topp = gr.Slider(0.5, 0.99, value=0.95, step=0.01, label="Top-P")

            generate_btn = gr.Button("Generate in my style", variant="primary")

        with gr.Column(scale=3):
            output = gr.Textbox(label="Your story continuation", lines=15)

    generate_btn.click(
        fn=generate_text,
        inputs=[prompt_input, max_new, temp, topk, topp],
        outputs=output
    )

print("✅ Gradio UI ready! Run the next cell to launch it.")
```
